# 3. Plot a bar graph of male vs female buyers participated in the sales

```python
car_mno = cars_data.loc[cars_data['Buyer Gender'] == 'Male'].count()
car_fno= cars_data.loc[cars_data['Buyer Gender'] == 'Female'].count()
print('No. of Male buyers =',car_mno['Buyer Gender'])
print('No. of Female buyers =',car_fno['Buyer Gender'])
```

```python
plt.figure(figsize=(4,7))
c_plot = sns.countplot(x='Buyer Gender', data=cars_data)
c_plot
plt.title('Male vs Female Buyers participated in the Sale',fontweight='bold',pad=30)
plt.ylim(4600,5200)
for p in c_plot.patches:
    c_plot.annotate(format(p.get_height(), '.2f'), 
                    (p.get_x() + p.get_width() / 2., 
                     p.get_height()), ha = 'center', va = 'center', size=12, 
                    xytext = (0, 10), textcoords = 'offset points')
```

### Insights

There were a total of 5052 male buyers and 4948 famale buyers participated in the sale

# 4. Find the top 5 cars based on their sales price

```python
cars_data.nlargest(5,['Sale Price'])
```

### Insights

The top 5 cars having the highest sales price are: Volkswagen Jetta III, Audi A6, Lexus LX, Mazda MX-6 and Lexus ES

# 5. Find the least 5 cars based on their Resell price

```python
cars_data.nsmallest(5,['Resell Price'])
```

### Insights

The 5 cars having the lowest Resell price are: Chevrolet 1500, GMC Savana 1500, Jaguar XF, Mistubishi 30000GT and Infiniti G37
