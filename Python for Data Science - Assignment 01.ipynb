# Visulaizations using the Dataset

```python
#count of each classification in the dataset
sns.countplot(x='Classification', data=iris_data)
```

```python
#Comapring the values of all the columns in the data set with each other
plt.figure(figsize=(12,12))
sns.pairplot(iris_data,hue='Classification')
```

```python
#Relation between SL and SW for each type of classificastion
plt.figure(figsize=(16,9))
plt.title("Relation between SL and SW of all the classifications")
sns.scatterplot(x='SL',y='SW',hue='Classification',data=iris_data,s=40)
```

```python
#Relation between PL and PW for each type of classificastion
plt.figure(figsize=(16,9))
plt.title("Relation between PL and PW of all the classifications")
sns.scatterplot(x='PL',y='PW',hue='Classification',data=iris_data,s=40)
```

```python
#correlation between different classifications
sns.heatmap(iris_data.corr(),annot=True)
plt.xticks()
plt.yticks(rotation=0)
```
